## 1. Check / install Java 17

## 2. Check / install Nextflow

```python
import os, shutil

if shutil.which("nextflow") is None:
    !curl -s https://get.nextflow.io | bash
    !chmod +x nextflow
    !mv nextflow /usr/local/bin/
!nextflow -version
```

## 3. Check / install Singularity / Apptainer

Colab does not ship with Singularity by default, so this step installs
**Apptainer** (the community continuation of Singularity, fully compatible).

```python
import shutil
if shutil.which("singularity") is None and shutil.which("apptainer") is None:
    !apt-get update -qq
    !apt-get install -y -qq software-properties-common
    !add-apt-repository -y ppa:apptainer/ppa
    !apt-get update -qq
    !apt-get install -y -qq apptainer
    !ln -sf $(which apptainer) /usr/local/bin/singularity
!singularity --version
```
